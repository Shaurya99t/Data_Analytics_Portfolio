```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

# 1
plt.subplot(2,2,1)
plt.title("Age")
plt.bar([1,2,3,4,5], [45,34,56,23,45])

# 2
plt.subplot(2,2,2)
plt.title("Weight")
plt.plot([5,6,7,8,9], [67,50,66,56,82])

# 3
plt.subplot(2,2,3)
plt.title("Weight")
plt.plot([5,6,7,8,9], [67,50,66,56,82])

# 4
plt.subplot(2,2,4)
plt.title("Weight")
plt.plot([5,6,7,8,9], [67,50,66,56,82])
plt.savefig("bar.png")

plt.tight_layout()
plt.show()
```
